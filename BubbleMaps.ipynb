# Bubble Maps

```python
import pandas as pd
import plotly.graph_objects as go

data=pd.read_csv('/home/kirti/Data_sets/district wise population and centroids.csv')
data=data.sort_values('Population in 2011',ascending=False)
data
```

```python
colors=["royalblue","crimson","lightseagreen","orange","lightgrey"]
limits=[(0,10),(10,50),(50,120),(120,250),(250,431)]
scale=50000
fig=go.Figure()

for i in range(len(limits)):
    lim=limits[i]
    df=data[lim[0]:lim[1]]
    fig.add_trace(go.Scattergeo
                  (geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
                  featureidkey='properties.ST_NM',
                  locationmode='geojson-id',
                  lon=df['Longitude'],
                  lat=df['Latitude'],
                  
                    marker=dict(
                      size=df['Population in 2011']/scale,
                      color=colors[i],
                      line_color='rgb(40,40,40)',
                      line_width=0.5,
                      sizemode = 'area')))
fig.update_geos(
         
         projection=dict(
                   type='conic conformal',
                    parallels=[12.4729,35.1728055],
                   rotation={'lat':24,'lon':80}),
         lonaxis={'range':[68,98]},
         lataxis={'range':[6,38]})
    
fig.show()    
```
