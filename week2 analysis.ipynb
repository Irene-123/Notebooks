```python
df=pd.read_csv('/home/kirti/Data_sets/presidents.csv')
df.head()
```

```python

df['First']=df['President']

df['First']=df['First'].replace('[ ].*','',regex=True)
df.head(10)
```

```python
del(df['First'])

def splitname(row):
    row['First']=row['President'].split(' ')[0]
    row['Last']=row['President'].split(' ')[-1]
    return row
df=df.apply(splitname, axis='columns')
df.head()
```

```python
del(df['First'])
del(df['Last'])
pattern='(^[\w]*)(?:.* )([\w]*$)'
df['President'].str.extract(pattern).head()
```

```python
df['Born']=df['Born'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df['Born'].head()
```

```python
df['Born']=pd.to_datetime(df['Born'])
df['Born'].head()
```

```python
df[['age1','end2']] = df['Age atend of presidency'].str.split("days", expand=True)
df['age1'] += 'days'
df.head()
```
