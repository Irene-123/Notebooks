```python
cat_totals=df.groupby("Major_category")["Total"].sum().sort_values()
```

```python
cat_totals.plot(kind='barh',fontsize=4)
```

```python
small_cat=cat_totals[cat_totals<100_000]
big_cat=cat_totals[cat_totals>100_000]
small_sums=pd.Series([small_cat.sum()],index=['other'])
big_cat=big_cat.append(small_sums)
big_cat.plot(kind="pie",label="")
```
